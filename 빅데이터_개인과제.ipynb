3.1 스크레이핑 절차

1. 키워드 설정: 위 주제 설정 및 관심 배경에서 짧게 상술했듯이, 스크레이핑을 위한 주요 키워드로는  ‘러시아-우크라이나 전쟁’, ‘러시아 전쟁’, ‘대러(한러) 관계’, 우크라이나 침공’,’푸틴’ 등을 설정하였다. 이 키워드들은 우크라이나-러시아 전쟁과 그로 인한 한러 관계와 관련된 핵심 이슈를 포괄한다.
2. URL 추출 및 데이터 수집: 연합 뉴스에서 각 키워드로 기사를 검색한 후, 해당 기사의 URL을 추출하였다. 추출된 URL을 바탕으로 기사이 제목, 발행 날짜, 본문을 html_nodes()와 html_text() 함수를 사용하여 수집하였다. 수집된 데이터는 Excel 파일로 저장되어 이후 분석에 사용될 기초 자료로 활용되었다.
3. 오류 처리 및 윤리적 고려 사항 : 스크레이핑 작업 중 발생할 수 있는 오류를 처리하기 위해 tryCatch 구문을 사용하였다. 또한, Sys.sleep(5)를 추가하여 요청 사이에 5초 지연 추가하여 서버 리소스를 존중하도록 하였다.

본 연구에서 수집된 500개의 기사는 향후 정량적 분석을 통해 러시아-우크라이나 전쟁 이후 한국 국민의 대러 인식을 심층적으로 분석할 중요한 기초 자료로 사용될 예정이다. 정성적 자료를 정량적 데이터로 변환하는 과정은 다음과 같이 진행된다.

```
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup

# 사용자 에이전트 설정
ua_string = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/540.0 (KHTML, like Gecko) Chrome/95.0.4606.81 Safari/540.0"
headers = {"User-Agent": ua_string}

# 파일 경로에서 Excel 읽기 (Colab에서 파일 업로드 후 사용)
file_path = "/content/drive/MyDrive/빅데이터_개인과제/연합뉴스_url.xlsx"  # Excel 파일 경로
news_source = pd.read_excel(file_path)

# 데이터 프레임에 'date', 'title', 'text' 열 추가
news_source["date"] = ""
news_source["title"] = ""
news_source["text"] = ""

# URL 목록을 반복하면서 스크래핑 수행
for i, row in news_source.iterrows():
    url = row["url"]
    try:
        # 요청 사이에 5초 지연
        time.sleep(5)

        # HTML 요청 및 파싱
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")

        # 데이터 추출
        temp_news_date = soup.select_one(".info-con")
        temp_news_title = soup.select_one("h1")
        temp_news_content = soup.select("p")

        # 추출된 데이터 저장 (없을 경우 None 처리)
        news_source.at[i, "date"] = temp_news_date.get_text(strip=True) if temp_news_date else None
        news_source.at[i, "title"] = temp_news_title.get_text(strip=True) if temp_news_title else None
        news_source.at[i, "text"] = " ".join([p.get_text(strip=True) for p in temp_news_content]) if temp_news_content else None

        # 로그 출력 (옵션)
        print(f"Scraped data from: {url}")

    except Exception as e:
        print(f"Error scraping {url}: {e}")
```

```
# 결과 저장
output_file_path = "/content/drive/MyDrive/빅데이터_개인과제/results.xlsx"  # 저장할 Excel 파일 경로
news_source.to_excel(output_file_path, index=False)

print("Scraping completed and saved to Excel.")
```

#3.데이터 분석

직접적인 분석을 들어가기 전, 기사 보도의 텍스트를 이용하여 단어의 빈도 분석, 워크 클라우드, LDA를 진행해보았다.

```
#======================================================
# 필요한 라이브러리 로딩
#======================================================
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import networkx as nx
from collections import Counter
import seaborn as sns
from gensim import corpora, models

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
#======================================================
# 데이터 파일 로딩
#======================================================
# Excel 파일 로딩
source_file = pd.read_excel("/content/drive/MyDrive/빅데이터_개인과제/results.xlsx")  # Excel 파일 경로 설정
text_source = source_file.copy()  # 데이터 복사
#======================================================
# 데이터 필터링 및 전처리
#======================================================
# 텍스트 필드 데이터 가져오기
text_data = text_source['text']
#======================================================
# 불용어 제거 및 토큰화
#======================================================
# 사용자 정의 불용어 추가
custom_stopwords = pd.read_excel("/content/drive/MyDrive/빅데이터_개인과제/English_Stopwords.xlsx")['English_Stopwords'].tolist()
nltk_stopwords = set(stopwords.words('english'))
combined_stopwords = set(custom_stopwords).union(nltk_stopwords)

# 토큰화 및 전처리 함수 정의
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalpha()]  # 알파벳만 유지
    tokens = [word for word in tokens if word not in combined_stopwords]  # 불용어 제거
    return tokens

text_tokens = text_data.apply(preprocess_text)

#======================================================
# 문서-용어 행렬 생성
#======================================================
vectorizer = CountVectorizer(tokenizer=lambda doc: doc, lowercase=False)
doc_term_matrix = vectorizer.fit_transform(text_tokens)
doc_term_matrix_df = pd.DataFrame(doc_term_matrix.toarray(), columns=vectorizer.get_feature_names_out())

#======================================================
# 빈도 분석
#======================================================
# 단어 빈도 계산
word_freq = doc_term_matrix_df.sum(axis=0).sort_values(ascending=False)
top_words = word_freq.head(50)

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
sns.barplot(x=top_words.values, y=top_words.index, palette='viridis')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title('Top 50 Words by Frequency')
plt.show()

#======================================================
# 워드 클라우드 생성
#======================================================
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud')
plt.show()

#======================================================
# 네트워크 분석
#======================================================
# 단어 공기어 행렬 생성
co_occurrence_matrix = (doc_term_matrix.T @ doc_term_matrix)
co_occurrence_df = pd.DataFrame(co_occurrence_matrix.toarray(), index=vectorizer.get_feature_names_out(), columns=vectorizer.get_feature_names_out())

# 상위 30개 단어로 네트워크 시각화
top_terms = word_freq.head(30).index
subgraph = co_occurrence_df.loc[top_terms, top_terms]

# 네트워크 그래프 생성
G = nx.from_pandas_adjacency(subgraph)
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue')
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=12)
plt.title('Word Co-Occurrence Network')
plt.show()

#======================================================
# 토픽 모델링
#======================================================
lda = LatentDirichletAllocation(n_components=3, random_state=42)
lda.fit(doc_term_matrix)

# 주요 토픽별 단어 추출
topic_words = {}
for topic, comp in enumerate(lda.components_):
    word_idx = np.argsort(comp)[::-1][:10]
    topic_words[f'Topic {topic + 1}'] = [vectorizer.get_feature_names_out()[i] for i in word_idx]

topic_words_df = pd.DataFrame(topic_words)
print(topic_words_df)
```

```
from transformers import pipeline
import pandas as pd

# 감정 분석 파이프라인 생성
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# 감정 분석 수행
def analyze_sentiment(text):
    try:
        result = sentiment_pipeline(text[:512])  # BERT 모델은 512 토큰 제한이 있음
        return result[0]['label'], result[0]['score']
    except Exception as e:
        return None, None

# 'text' 컬럼에서 감정 분석 수행
results_df['sentiment_label'], results_df['sentiment_score'] = zip(*results_df['text'].apply(analyze_sentiment))

# 'date' 컬럼에서 분기 정보 추출
results_df['quarter'] = results_df['date'].dt.to_period('Q')

# 분기별 평균 감정 점수 계산
quarterly_sentiment = results_df.groupby('quarter')['sentiment_score'].mean().reset_index()

# 결과 출력
print(quarterly_sentiment)

# 결과를 Excel 파일로 저장
output_file_path = "/content/drive/MyDrive/빅데이터_개인과제/quarterly_sentiment_results.xlsx"
quarterly_sentiment.to_excel(output_file_path, index=False)
```

수집된 기사를 분기별로 집계하여 감정 점수를 산출하였다. 그리하여, 시간에 따른 대러 감정의 변화를 시각화할 예정이다. 또한, 특정 사건 발생 시 전후의 감정 점수 변화를 분석하여 사건이 여론에 미친 영향을 평가할 계획이다.

#단순회귀 분석

```
import pandas as pd
import statsmodels.api as sm

# 분기별 병합 데이터 읽기
all_merged_data_path = "/content/drive/MyDrive/빅데이터_개인과제/분기별_병합데이터.xlsx"
all_merged_df = pd.read_excel(all_merged_data_path)

# 단순 회귀 분석 (감성지수 -> 러시아 호감도)
# 독립변수: Sentiment, 종속변수: RussiaFavorability
X_simple = all_merged_df['Sentiment']  # 독립변수
y_simple = all_merged_df['RussiaFavorability']  # 종속변수

# 상수항 추가 (회귀분석에서 필수)
X_simple = sm.add_constant(X_simple)

# 회귀 모델 적합
model_simple = sm.OLS(y_simple, X_simple).fit()

# 단순 회귀 분석 결과 출력
print("단순 회귀 분석 결과:")
print(model_simple.summary())

# 다중 회귀 분석 (감성지수 + 교역량 -> 러시아 호감도)
# 독립변수: Sentiment, 통제변수: TradeVolume, 종속변수: RussiaFavorability
X_multiple = all_merged_df[['Sentiment', 'TradeVolume']]  # 독립변수 + 통제변수
y_multiple = all_merged_df['RussiaFavorability']  # 종속변수

# 상수항 추가 (회귀분석에서 필수)
X_multiple = sm.add_constant(X_multiple)

# 회귀 모델 적합
model_multiple = sm.OLS(y_multiple, X_multiple).fit()

# 다중 회귀 분석 결과 출력
print("다중 회귀 분석 결과:")
print(model_multiple.summary())
```
